## Curvas parametrizadas

```python
from sympy import plot_parametric #Para graficar curvas parametrizadas importamos la libreria plot_ parametric
```

```python
s=symbols('s') # Defino a "s" como funci√≥n
```

```python
plot_parametric((cos(s)),sin(s)) #Comando para graficar una funci√≥n parametrica, en este caso las formulas parametricas
# para el circulo son cos(s) & sin(s)
```

```python
plot_parametric((cos(s),sin(s)),(s,s**2),(s,-4,4)) #Con este comando puede grafciar una curva parametrizada
# que es el circulo (cos(s),sin(s))
# Y adem√°s una funci√≥n normal  (s,s**2)
#Nota: Ambas curvas deben ser generadas por el mismo termino algebraico que se definio, en este caso es "s"
#Tambi√©n, podemos acotar el dominio de las funciones con (s,-4,4)
```

## Graficas 3D

```python
from sympy.plotting import plot3d #Importamos la libreria plot3d para graficas en 3 dimensiones.
```

```python
x,y=symbols('x y') # Comando para definir a "x" & "y" de forma simultanea como simolos algebraicos
```

```python
plot3d(x*y) #Coamdno para graficar en 3D
```

```python
plot3d(x*y, (x,-5,5),(y,-5,5)) #Comando para acotar los valores de los ejes x & y respectivamente.
```

```python
plot3d(x*y, -x*y, (x,-5,5),(y,-5,5)) #De la misma manera que las graficas en 2D, puedo anidar dos o mas graficas
                                    # para que visualicen en la misma imagen
```

```python
plot3d((x**2 + y**2, (x,-5,5),(y,-5,5)),(x*y,(x,-3,3),(y,-3,3)))
```

## Documentaci√≥n.
El comando $plot3d((x**2 + y**2, (x, -5, 5), (y, -5, 5)), (x*y, (x, -3, 3), (y, -3, 3)))$ se utiliza para generar dos gr√°ficas tridimensionales en una sola visualizaci√≥n. Cada comando genera lo siguiente:

Primera gr√°fica:

$x^2 + y^2$: Esta es la funci√≥n que se va a graficar, que representa una par√°bola en 3D.

$(x, -5, 5)$: Define el rango de valores para la variable x, desde -5 hasta 5.

$(y, -5, 5)$: Define el rango de valores para la variable y, desde -5 hasta 5.

Segunda gr√°fica:

$x*y$: Esta es la funci√≥n que se va a graficar, que representa una superficie en 3D donde la altura es el producto de x e y.

$(x, -3, 3)$: Define el rango de valores para la variable x, desde -3 hasta 3.

$(y, -3, 3)$: Define el rango de valores para la variable y, desde -3 hasta 3.

```python
from sympy.plotting import plot3d_parametric_line
```

## Documentaci√≥n.
La librer√≠a plot3d_parametric_line se utiliza para generar gr√°ficas tridimensionales de l√≠neas param√©tricas.

Esta funci√≥n permite visualizar curvas en el espacio 3D que est√°n definidas por par√°metros.

## Documentaci√≤n.

La librer√≠a plot3d_parametric_surface de SymPy se utiliza para generar gr√°ficas tridimensionales de superficies param√©tricas. 

Esta funci√≥n permite visualizar superficies en el espacio 3D que est√°n definidas por dos par√°metros.

## Funciones impl√≠citas

```python
from sympy import plot_implicit, symbols, Eq, And
```

## Documentaci√≤n

$plot_implicit:$ Esta funci√≥n se usa para graficar ecuaciones impl√≠citas en un plano. Por ejemplo, si tienes una ecuaci√≥n que relaciona x y y, puedes visualizar la curva que representa esa relaci√≥n.

$symbols:$ Esta funci√≥n se utiliza para definir s√≠mbolos matem√°ticos que puedes usar en tus ecuaciones. Por ejemplo, puedes crear s√≠mbolos para x e y.

$Eq:$ Esta clase se utiliza para definir ecuaciones de manera simb√≥lica. Permite establecer relaciones entre diferentes expresiones.

$And:$ Esta funci√≥n se usa para combinar m√∫ltiples condiciones o ecuaciones. Puede ser √∫til si deseas graficar la intersecci√≥n de varias ecuaciones.

```python
x, y = symbols(' x y') #Defino mis variables algebraicas
```

```python
p1 = plot_implicit(Eq(x**2 + y**2, 5)) #Comando para graficar una funcion implicita, en este caso la ec. general de la cirfunferencia con radio sqrt(5)
```

```python
p6 = plot_implicit(y > x**2) # Este comando grafica la condici√≥n que se desea saber de una grafica, por ejemplo,
# en este caso se cumple la regi√≥n del plano donde la variable "y" es mayor que la expresi√≥n x^2
```

```python
p6 = plot_implicit(And(y > x, y > -x)) # La funci√≥n que realiza el comando And(y > x, y > -x) es generar la intersecci√≥n
# de ambas condiciones que se deben cumplir.
```
