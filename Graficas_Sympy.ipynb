## Documentación.
La función plot(x**2, adaptative=False, nb_of_points=10)

Se utiliza para generar una gráfica de la función con los siguientes parámetros:

$x^2$: Define la función que se va a graficar, en este caso ( y = x^2 ).

$adaptative=False$: Indica que la gráfica no debe ajustar automáticamente el número de puntos en función de la complejidad de la función.

$nb$ $of$ $points=10$: Especifica que se deben usar exactamente 10 puntos para crear la gráfica.

Esto significa que la gráfica se generará con 10 puntos distribuidos uniformemente a lo largo del rango de x, sin adaptarse a la forma de la función.

## Graficas 3D

```python
from sympy.plotting import plot3d #Importamos la libreria plot3d para graficas en 3 dimensiones.
```

```python
x,y=symbols('x y') # Comando para definir a "x" & "y" de forma simultanea como simolos algebraicos
```

```python
plot3d(x*y) #Coamdno para graficar en 3D
```

```python
plot3d(x*y, (x,-5,5),(y,-5,5)) #Comando para acotar los valores de los ejes x & y respectivamente.
```

```python
plot3d(x*y, -x*y, (x,-5,5),(y,-5,5)) #De la misma manera que las graficas en 2D, puedo anidar dos o mas graficas
                                    # para que visualicen en la misma imagen
```

```python
plot3d((x**2 + y**2, (x,-5,5),(y,-5,5)),(x*y,(x,-3,3),(y,-3,3)))
```

## Documentación.
El comando $plot3d((x**2 + y**2, (x, -5, 5), (y, -5, 5)), (x*y, (x, -3, 3), (y, -3, 3)))$ se utiliza para generar dos gráficas tridimensionales en una sola visualización. Cada comando genera lo siguiente:

Primera gráfica:

$x^2 + y^2$: Esta es la función que se va a graficar, que representa una parábola en 3D.

$(x, -5, 5)$: Define el rango de valores para la variable x, desde -5 hasta 5.

$(y, -5, 5)$: Define el rango de valores para la variable y, desde -5 hasta 5.

Segunda gráfica:

$x*y$: Esta es la función que se va a graficar, que representa una superficie en 3D donde la altura es el producto de x e y.

$(x, -3, 3)$: Define el rango de valores para la variable x, desde -3 hasta 3.

$(y, -3, 3)$: Define el rango de valores para la variable y, desde -3 hasta 3.

```python
from sympy.plotting import plot3d_parametric_line
```

## Documentación.
La librería plot3d_parametric_line se utiliza para generar gráficas tridimensionales de líneas paramétricas.

Esta función permite visualizar curvas en el espacio 3D que están definidas por parámetros.

## Funciones implícitas

```python
from sympy import plot_implicit, symbols, Eq, And
```

## Documentaciòn

$plot_implicit:$ Esta función se usa para graficar ecuaciones implícitas en un plano. Por ejemplo, si tienes una ecuación que relaciona x y y, puedes visualizar la curva que representa esa relación.

$symbols:$ Esta función se utiliza para definir símbolos matemáticos que puedes usar en tus ecuaciones. Por ejemplo, puedes crear símbolos para x e y.

$Eq:$ Esta clase se utiliza para definir ecuaciones de manera simbólica. Permite establecer relaciones entre diferentes expresiones.

$And:$ Esta función se usa para combinar múltiples condiciones o ecuaciones. Puede ser útil si deseas graficar la intersección de varias ecuaciones.

```python
x, y = symbols(' x y') #Defino mis variables algebraicas
```

```python
p1 = plot_implicit(Eq(x**2 + y**2, 5)) #Comando para graficar una funcion implicita, en este caso la ec. general de la cirfunferencia con radio sqrt(5)
```

```python
p6 = plot_implicit(y > x**2) # Este comando grafica la condición que se desea saber de una grafica, por ejemplo,
# en este caso se cumple la región del plano donde la variable "y" es mayor que la expresión x^2
```

```python
p6 = plot_implicit(And(y > x, y > -x)) # La función que realiza el comando And(y > x, y > -x) es generar la intersección
# de ambas condiciones que se deben cumplir.
```
